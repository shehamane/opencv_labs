```python
import cv2
import numpy as np
```

```python
def nothing(x):
    pass

cap = cv2.VideoCapture('car.mp4')
cv2.namedWindow('video')
cv2.namedWindow('video_edges')

cv2.createTrackbar('min', 'video_edges', 0, 255, nothing)
cv2.createTrackbar('max', 'video_edges', 0, 255, nothing)


while True:
    ret, frame = cap.read()
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    min_value = cv2.getTrackbarPos('min', 'video_edges')
    max_value = cv2.getTrackbarPos('max', 'video_edges')

    frame_x = cv2.Sobel(gray_frame, cv2.CV_64F, 1, 0, ksize = 5)
    frame_y = cv2.Sobel(gray_frame, cv2.CV_64F, 0, 1, ksize = 5)
    edge_frame = cv2.Canny(gray_frame, min_value, max_value)

    # cv2.imshow('video_x', frame_x)
    # cv2.imshow('video_y', frame_y)
    cv2.imshow('video', frame)
    cv2.imshow('video_edges', edge_frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
