```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture('car.mp4')

cv2.namedWindow('tracks')
cv2.createTrackbar('lower', 'tracks', 0, 255, lambda x: None)
cv2.createTrackbar('upper', 'tracks', 0, 255, lambda x: None)

while True:
    ret, frame = cap.read()

    if ret:
        frame = cv2.bilateralFilter(frame, 9, 75, 75)
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        canny_frame = cv2.Canny(gray_frame,
                                cv2.getTrackbarPos('lower', 'tracks'),
                                cv2.getTrackbarPos('upper', 'tracks'))
        dilate_frame = cv2.dilate(canny_frame, np.ones((5, 5)), iterations=1)

        contours, _ = cv2.findContours(dilate_frame, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
        contours = sorted(contours, key=cv2.contourArea, reverse=True)

        for contour in contours:
            area = cv2.contourArea(contour)
            if area > 100:
                cv2.drawContours(frame, contour, -1, (0, 255, 0), 3)
                p = cv2.arcLength(contour, True)
                verts = cv2.approxPolyDP(contour,0.01*p, True)
                x, y, w, h = cv2.boundingRect(verts)
                cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 256), 4)

        cv2.imshow('original', frame)
        cv2.imshow('canny', canny_frame)
        cv2.imshow('dilate', dilate_frame)
    else:
        cap.set(cv2.CAP_PROP_POS_FRAMES, 0)
        continue

    if cv2.waitKey(1) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
