```python
import cv2
import numpy as np
import time
```

```python
cap = cv2.VideoCapture('car.mp4')

while True:
    ret, frame = cap.read()

    blur_frame = cv2.blur(frame, (11, 11))
    gblur_frame = cv2.GaussianBlur(frame, (11, 11), 0)
    bblur_frame = cv2.bilateralFilter(frame, 9, 75, 75)

    cv2.imshow('video', frame)
    cv2.imshow('blur_video', blur_frame)
    cv2.imshow('gblur_video', gblur_frame)
    cv2.imshow('bblur_video', bblur_frame)

    if cv2.waitKey(1) & 0xFF == 27:
            break

cv2.destroyAllWindows()
```

```python
def nothing(x):
    pass

kernel = np.ones(shape=(5, 5), dtype=np.uint8)

cap = cv2.VideoCapture('car.mp4')
cv2.namedWindow('mask')


cv2.createTrackbar("HL", "mask", 0, 180, nothing)
cv2.createTrackbar("SL", "mask", 0, 255, nothing)
cv2.createTrackbar("VL", "mask", 0, 255, nothing)

cv2.createTrackbar("H", "mask", 0, 180, nothing)
cv2.createTrackbar("S", "mask", 0, 255, nothing)
cv2.createTrackbar("V", "mask", 0, 255, nothing)


while True:
    ret, frame = cap.read()

    if ret:
        frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

        hl = cv2.getTrackbarPos("HL", "mask")
        sl = cv2.getTrackbarPos("SL", "mask")
        vl = cv2.getTrackbarPos("VL", "mask")

        h = cv2.getTrackbarPos("H", "mask")
        s = cv2.getTrackbarPos("S", "mask")
        v = cv2.getTrackbarPos("V", "mask")

        lower = np.array([hl, sl, vl])
        upper = np.array([h, s, v])
        frame_blured = cv2.bilateralFilter(frame, 9, 75, 75)
        mask = cv2.inRange(frame_hsv, lower, upper)
        res = cv2.bitwise_and(frame_blured, frame_blured, mask=mask)
        opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
        closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)
        edge = cv2.Canny(mask, 100, 200)

        contours, _ = cv2.findContours(edge, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
        contours = sorted(contours, key = cv2.contourArea, reverse=True)

        if contours:
            cv2.drawContours(frame, [contours[0]], -1, (255, 0, 0), 5)

        cv2.imshow('video', frame)
        cv2.imshow('closing', closing)
    else:
        cap.set(cv2.CAP_PROP_POS_FRAMES, 0)
        continue
    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
